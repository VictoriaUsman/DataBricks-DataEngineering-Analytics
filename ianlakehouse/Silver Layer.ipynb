### Reading and Cleaning The Tables

```python
from pyspark.sql.functions import when, col
```

```python
%sql
select * from iandatabricks.bronze.crm_custinfo
```

Trimming the customer's first and lastname

```python
from pyspark.sql.functions import trim, ltrim, rtrim
df = spark.sql("select * from iandatabricks.bronze.crm_custinfo")
df = df.withColumn("cst_firstname", trim(df["cst_firstname"]))
trimmed_df = df.withColumn("cst_lastname", trim(df["cst_lastname"]))
```

```python
display(trimmed_df)
```

```python
trans_df = trimmed_df.withColumn(
    "cst_marital_status",
    when(trim(col("cst_marital_status")) == "S", "Single")
    .when(trim(col("cst_marital_status")) == "M", "Married")
)
```

```python
trans_df = trans_df.withColumn("cst_gndr",
                               when(trim(col("cst_gndr")) == "M", "Male")
                             .when(trim(col("cst_gndr")) == "F", "Female"))
```
